## 1. Desarrollo de Algoritmos Secuenciales
Se implementaron los algoritmos **MergeSort** y **QuickSort** en su versión secuencial.

El diseño modular del código permitirá paralelizar fácilmente en OpenMP y CUDA en siguientes etapas.

````
// QuickSort Secuencial en C++
void quickSort(std::vector<int>& arr, int left, int right) {
    int i = left, j = right;
    int pivot = arr[(left + right) / 2];

    while (i <= j) {
        while (arr[i] < pivot) i++;
        while (arr[j] > pivot) j--;
        if (i <= j) {
            std::swap(arr[i], arr[j]);
            i++;
            j--;
        }
    }

    if (left < j) quickSort(arr, left, j);
    if (i < right) quickSort(arr, i, right);
}
````

````
// MergeSort Secuencial en C++
void merge(std::vector<int>& arr, int l, int m, int r) {
    int n1 = m - l + 1;
    int n2 = r - m;
    std::vector<int> L(n1), R(n2);
    for (int i = 0; i < n1; i++) L[i] = arr[l + i];
    for (int j = 0; j < n2; j++) R[j] = arr[m + 1 + j];

    int i = 0, j = 0, k = l;
    while (i < n1 && j < n2) {
        if (L[i] <= R[j]) arr[k++] = L[i++];
        else arr[k++] = R[j++];
    }
    while (i < n1) arr[k++] = L[i++];
    while (j < n2) arr[k++] = R[j++];
}

void mergeSort(std::vector<int>& arr, int l, int r) {
    if (l < r) {
        int m = l + (r - l) / 2;
        mergeSort(arr, l, m);
        mergeSort(arr, m + 1, r);
        merge(arr, l, m, r);
    }
}
````

## 2. Validación de Implementaciones
Se generaron datos aleatorios y se validó la salida de los algoritmos comparándola con `std::sort` de la STL de C++.

**Casos de prueba:**
- Arreglos de tamaño $10^3$, $10^5$, $10^6$ elementos.
- Verificación de orden correcto.
- Cálculo de tiempos de ejecución iniciales.

````
// Generación de datos aleatorios
std::vector<int> generateRandomArray(int size) {
    std::vector<int> arr(size);
    std::generate(arr.begin(), arr.end(), [](){ return rand() % 10000; });
    return arr;
}
````

**Ejemplo de validación:**
```cpp
auto arr = generateRandomArray(10000);
auto arr_copy = arr;
quickSort(arr, 0, arr.size() - 1);
std::sort(arr_copy.begin(), arr_copy.end());
assert(arr == arr_copy);
```
